```python
#importing libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,classification_report,roc_auc_score
from sklearn.metrics import precision_score,recall_score
```

```python
#loading CSV file
data=pd.read_csv('creditcard.csv')
```

```python
#records which have non-fraud transcations
data_with_non_fraud=data[data['Class']==0]
data_with_non_fraud.shape
```

```python
#records which have fraud transcations
data_with_fraud=data[data['Class']==1]
data_with_fraud.shape
```

```python
#checking if there are any missing values
print(data.isnull().sum())
```

```python
#checking duplicates in the dataset
print(data.duplicated().sum())
```

```python
#dropping all the duplicates
data = data.drop_duplicates()
data.shape
```

```python
#checking data types of the data
print(data.dtypes)
```

```python
x=data.drop('Class', axis=1)
y=data['Class']
```

```python
#splitting the whole data into 80% training data and 20% testing data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=40,stratify=y)
```

```python
#Baseline model selection
#Selected Standard Scaler as the data is uneven(some values are low and some values are high)
#So to make them equal we use StandardScaler
scaler=StandardScaler()
```

```python
#Fitting model
#Here we have only used x_train as we the test data learns information from the future which is data leakage
scaler.fit(x_train)
```

```python
#Transfrom both training and test data
x_train_scaled_data=scaler.transform(x_train)
x_test_scaled_data=scaler.transform(x_test)
```

```python
#Model Selection
model=LogisticRegression(class_weight='balanced', random_state=42, max_iter=1000)
```

```python
#Fitting model
model.fit(x_train_scaled_data,y_train)
```

```python
#Predicting
y_pred=model.predict(x_test_scaled_data)
y_pred
```

```python
#Confusion mactrix
print(confusion_matrix(y_test, y_pred))
```

```python
#Classification Report(gives precision, recall, F1-score)
print(classification_report(y_test, y_pred))
```

```python
#ROC-AUC score
print(roc_auc_score(y_test, y_pred))
```

```python
#Probabilities of predictions for threhold tuning
y_pred_proba=model.predict_proba(x_test_scaled_data)[:,1]
y_pred_proba
```

```python
#Threshold tuning
thresholds = [0.50, 0.30, 0.20, 0.10, 0.05]
for t in thresholds:
    print(f'Threshold: {t}') 
    y_pred_t=(y_pred_proba>=t).astype(int)
    #Confusion matrix
    cm = confusion_matrix(y_test, y_pred_t)
    print("Confusion Matrix:\n", cm)
    #precision 
    precision = precision_score(y_test, y_pred_t)
    print('Precision:', precision)
    #recall
    recall = recall_score(y_test, y_pred_t)
    print('Recall: ', recall)
```

```python
#Here from above, for threshold 0.05, recall is 0.96, it means 96% catches fraud
#Best threshold value now
low_thresholds = np.arange(0.001, 0.11, 0.001)
#finding best recall
best_recall=0
best_threshold=None
best_cm=None
for t in low_thresholds:
    y_pred_t=(y_pred_proba>=t).astype(int)
    #Confusion matrix
    cm = confusion_matrix(y_test, y_pred_t)
    #precision 
    precision = precision_score(y_test, y_pred_t)
    #recall
    recall = recall_score(y_test, y_pred_t)
    if recall>best_recall:
        best_recall=recall
        best_threshold=t
        best_cm=cm = confusion_matrix(y_test, y_pred_t)

print("Best Threshold for max Recall:", best_threshold)
print("Highest Recall:", best_recall)
print("Confusion Matrix at Best Threshold:\n", best_cm)   
```
